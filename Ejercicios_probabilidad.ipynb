```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
# # \*\* 1. Ejemplo de un tirador\*\*

## **Problema**\_un tirador debe acertar dentro de un circulo de radio 2 contenido en una region de lado 12. ¿Cual es la probabilidad de que el tirador acierte?

# Parámetros
s = 12  # lado del cuadrado
r = 2   # radio del círculo

# Áreas
area_cuadrado = s ** 2
area_circulo = np.pi * r ** 2
probabilidad = area_circulo / area_cuadrado

# Visualización
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(0, s)
ax.set_ylim(0, s)
ax.set_title('Probabilidad Geométrica: Punto en un Área')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Dibujar cuadrado
square = patches.Rectangle((0, 0), s, s, edgecolor='black', facecolor='none', linewidth=2)
ax.add_patch(square)

# Dibujar círculo
circle = patches.Circle((s / 2, s / 2), r, edgecolor='red', facecolor='none', linewidth=2)
ax.add_patch(circle)

# Agregar texto
ax.text(s / 2, s * 0.9, f"Área Cuadrado = {area_cuadrado}", ha='center', fontsize=10)
ax.text(s / 2, s * 0.1, f"Área Círculo = {area_circulo:.2f}", ha='center', fontsize=10, color='red')
ax.text(s / 2, s * 0.5, f"Probabilidad = {probabilidad:.4f}", ha='center', fontsize=11, color='blue')

plt.grid(True)
plt.show()

# Imprimir en consola

```

**Probabilidad de encuentro de amigos **

 ## Problema

Dos amigos acuerdan encontrarse en un lugar entre las **12:00 PM y 1:00 PM**.
Cada uno llega en un momento **aleatorio y uniforme** dentro de esa hora.
El primero que llega **espera 15 minutos**, luego se va si el otro no ha llegado.
Queremos encontrar la **probabilidad de que se encuentren**.

---

## Solución

Sea:

* \$X\$: la hora (en minutos después de las 12:00) en que llega el primer amigo.
* \$Y\$: la hora en que llega el segundo amigo.

Ambos \$X\$ y \$Y\$ son variables aleatorias uniformemente distribuidas en $\[0, 60]\$.

Se encontrarán si la diferencia de tiempo entre ellos es **menor o igual a 15 minutos**:

$$|X - Y| \leq 15$$

Esto define una región en el cuadrado de \$60 \times 60\$ minutos donde se encuentran.
Visualmente, es el área entre las rectas:

$$Y = X + 15 \quad \text{y} \quad Y = X - 15$$

---

## Cálculo geométrico

**El área total del cuadrado es:**

$$
A_{\text{total}} = 60 \times 60 = 3600
$$

La región donde **no se encuentran** (es decir, cuando \$|X - Y| > 15\$) está fuera de las bandas anteriores, y consiste en **dos triángulos** de lado 45:

$$
A_{\text{no encuentro}} = 2 \times \frac{1}{2} \times 45 \times 45 = 2025
$$

Por lo tanto, el área donde **sí se encuentran** es:

$$
A_{\text{encuentro}} = 3600 - 2025 = 1575
$$

---

## Resultado final

**La probabilidad de que se encuentren es:**

$$
P = \frac{A_{\text{encuentro}}}{A_{\text{total}}} = \frac{1575}{3600} = \frac{7}{16} \approx 0.4375
$$

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Parámetros
t_max = 60     # Tiempo total en minutos
espera = 15    # Tiempo máximo de espera

# Crear figura y ejes
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(0, t_max)
ax.set_ylim(0, t_max)
ax.set_xlabel("Llegada del amigo 1 (min)")
ax.set_ylabel("Llegada del amigo 2 (min)")
ax.set_title("Región de Encuentro entre Dos Amigos")

# Coordenadas del polígono (región de encuentro)
x_poly = [0, espera, t_max, t_max, t_max - espera, 0]
y_poly = [espera, 0, t_max - espera, t_max, t_max, espera]

# Dibujar región de encuentro
ax.fill(x_poly, y_poly, color='lightblue', label='Región de encuentro')

# Dibujar líneas de frontera: Y = X + 15 y Y = X - 15
x_vals = np.array([0, t_max])
ax.plot(x_vals, x_vals + espera, linestyle='--', color='blue', label='Y = X + 15')
ax.plot(x_vals, x_vals - espera, linestyle='--', color='blue', label='Y = X - 15')

# Dibujar línea diagonal Y = X
ax.plot(x_vals, x_vals, color='black', linewidth=1.5, label='Y = X')

# Leyenda
ax.legend(loc='upper right')

# Cuadro de la figura
plt.grid(True)
plt.show()

```
